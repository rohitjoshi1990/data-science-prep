```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### 1. Beta distribution

```python
beta = np.random.beta(1,2,size=1000)

# Add a vertical line at the mean.
plt.axvline(beta.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(beta.mean() + beta.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(beta.mean()-beta.std(), color='b', linestyle='dashed', linewidth=2) 

plt.hist(beta)
plt.show()
```

#### This distribution looks skewed towards the left. The mean is not the measure of central tendency. Also the std deviation does not cover same number on elements on both the sides of the mean.

### 2. Chisquare distribution

```python
chisquare = np.random.chisquare(1,size=1000)

# Add a vertical line at the mean.
plt.axvline(chisquare.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(chisquare.mean() + chisquare.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(chisquare.mean()-chisquare.std(), color='b', linestyle='dashed', linewidth=2) 


plt.hist(chisquare)
plt.show()
```

#### This distribution looks skewed towards the left. The mean is not the measure of central tendency. Also the std deviation does not cover same number on elements on both the sides of the mean.

### 3. Drichlet distribution

```python
drichlet = np.random.dirichlet([1,2,3,4],size=1000)

# Add a vertical line at the mean.
plt.axvline(drichlet.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(drichlet.mean() + drichlet.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(drichlet.mean()-drichlet.std(), color='b', linestyle='dashed', linewidth=2) 

plt.hist(drichlet)
plt.show()
```

#### This distribution looks skewed towards the left. The mean is not the measure of central tendency. Median might be a good one. Also the std deviation does not cover same number on elements on both the sides of the mean. However, it covers with a better balance than previously discussed distributions.

### 4. Exponential

```python
exp = np.random.exponential(scale=1,size=1000)

# Add a vertical line at the mean.
plt.axvline(exp.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(exp.mean() + exp.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(exp.mean()-exp.std(), color='b', linestyle='dashed', linewidth=2) 

plt.hist(exp)
plt.show()
```

#### This distribution looks skewed towards the left.

### 5. f distribution 

```python
f = np.random.f(50,100,size=1000)

# Add a vertical line at the mean.
plt.axvline(f.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(f.mean() + f.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(f.mean()-f.std(), color='b', linestyle='dashed', linewidth=2) 

plt.hist(f)
plt.show()
```
